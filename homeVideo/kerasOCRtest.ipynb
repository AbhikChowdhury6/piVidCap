```python
import cv2
import keras_ocr

pipeline = keras_ocr.pipeline.Pipeline()

testFile = '/media/chowder/abhiksFiles/personalData/wyzeCams/roomCamBy3dp/20201230/10/20.mp4'

cap = cv2.VideoCapture(testFile)
frames = []
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    x1, y1 = 1350, 1000  # Starting coordinates
    x2, y2 = 1900, 1080  # Ending coordinates
    cropped_frame = frame[y1:y2, x1:x2]
    frames.append(cropped_frame)

```

```python
#26 sec
prediction_groups = pipeline.recognize(frames[:60])
```

```python
#42 sec
prediction_groups = pipeline.recognize(frames[:100])
```

```python
prediction_groups = pipeline.recognize(frames[:200])
```
